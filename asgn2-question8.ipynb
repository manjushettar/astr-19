```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)
```

```python
def sin(x):
    return np.sin(x)
```

```python
def cos(x):
    return np.cos(x)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].plot(x, sin(x))
axs[0].set_title('sin(x)')

axs[1].plot(x, cos(x))
axs[1].set_title('cos(x)')

plt.savefig("multipanel_plot.pdf")
```
