```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.normal(0,1,1000)
```

```python
plt.figure()
plt.hist(x, bins = 100, density = True, edgecolor = 'black')
plt.xlabel("Random Number")
plt.ylabel("Frequency")
plt.title("Gaussian Distributed Random Numbers")
```
